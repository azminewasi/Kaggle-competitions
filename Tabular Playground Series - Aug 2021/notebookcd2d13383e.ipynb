```python
train=pd.read_csv("/kaggle/input/tabular-playground-series-aug-2021/train.csv")
test=pd.read_csv("/kaggle/input/tabular-playground-series-aug-2021/test.csv")
```

```python
train.head()
```

```python
test=test.set_index('id')
test.head()
```

```python
train=train.drop('id', 1)
y=train['loss']
X=train.drop('loss',1)
```

```python

from sklearn import preprocessing
import xgboost as xgb
from xgboost.sklearn import XGBRegressor
import datetime
from sklearn.model_selection import GridSearchCV,KFold, cross_validate, train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25)
eval_set=[(X_test,y_test)]
```

```python
best=XGBRegressor(objective='reg:squarederror',
              learning_rate=.6, 
              n_estimators=500)
best.fit(X,y,eval_set= eval_set, eval_metric= "rmse",early_stopping_rounds= 150,verbose=True)
```

```python
preds = best.predict(test)
submission = pd.DataFrame({"id": test.index, "loss": preds})
submission.to_csv("submission.csv", index=False)
print("Done")
```
